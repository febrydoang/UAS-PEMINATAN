```python
print('\n==========Program Kasir Sederhana==========')

pembeli = input('Masukkan Nama Pembeli: ')

print("Nama Pembeli:", pembeli)

def fungsimakanan():
    global totalmkn
    global porsi
    global mkn
    print("\n==========Menu Makanan==========\n")
    print("1. Bakso-----Rp. 10000")
    print("2. Mie Ayam-----Rp. 13000")
    print("3. Soto-----Rp. 12000")
    print("4. Nasi Goreng-----Rp. 15000")
    print("5. Kwetiau-----Rp. 12000")
    nomor = int(input("\nMasukan Pilihan: "))
    porsi = int(input("Berapa Porsi: "))
    if nomor == 1:
        totalmkn = porsi * 10000
        print(porsi, "Porsi Bakso = Rp", totalmkn)
        mkn = "Bakso"
    elif nomor == 2:
        totalmkn = porsi * 13000
        print(porsi, "Porsi Mie Ayam = Rp", totalmkn)
        mkn = "Mie Ayam"
    elif nomor == 3:
        totalmkn = porsi * 12000
        print(porsi, "Porsi Soto = Rp", totalmkn)
        mkn = "Soto"
    elif nomor == 4:
        totalmkn = porsi * 15000
        print(porsi, "Porsi Nasi Goreng = Rp", totalmkn)
        mkn = "Nasi Goreng"
    elif nomor == 5:
        totalmkn = porsi * 12000
        print(porsi, "Porsi Kwetiau = Rp", totalmkn)
        mkn = "Kwetiau"
    else:
        print("Pilihan tidak ada. Silahkan Masukan Lagi!!!")
        fungsimakanan()

def fungsiminuman():
    global totalmnm
    global mnm
    global gelas
    print("\n==========Menu Minuman==========\n")
    print("1. Air Mineral ---Rp 3000")
    print("2. Es Teh Manis ---Rp 4000")
    print("3. Es Jeruk ---Rp 7000")
    print("4. Es Kopi ---Rp 8000")
    print("5. Es Nutrisari ---Rp 5000")
    nomor = int(input("\nMasukan Pilihan Minuman: "))
    gelas = int(input("Berapa Gelas: "))
    if nomor == 1:
        totalmnm = gelas * 3000
        print(gelas, "Gelas Air Mineral = Rp", totalmnm)
        mnm = "Air Mineral"
    elif nomor == 2:
        totalmnm = gelas * 4000
        print(gelas, "Gelas Es Teh Manis = Rp", totalmnm)
        mnm = "Es Teh Manis"
    elif nomor == 3:
        totalmnm = gelas * 7000
        print(gelas, "Gelas Es Jeruk = Rp", totalmnm)
        mnm = "Es Jeruk"
    elif nomor == 4:
        totalmnm = gelas * 8000
        print(gelas, "Gelas Es Kopi = Rp", totalmnm)
        mnm = "Es Kopi"
    elif nomor == 5:
        totalmnm = gelas * 5000
        print(gelas, "Gelas Es Nutrisari = Rp", totalmnm)
        mnm = "Es Nutrisari"
    else:
        print("Pilihan tidak ada. Silahkan Masukan Lagi!!!")
        fungsiminuman()

def hitung_diskon(total):
    if total > 50000:
        diskon = total * 0.1
        print("\nSelamat! Anda mendapatkan diskon 10% sebesar: Rp", int(diskon))
        total -= diskon
    else:
        diskon = 0
    return int(total)

def menu_ulang():
    global totalmkn
    global totalmnm
    totalmkn = 0
    totalmnm = 0
    while True:
        fungsimakanan()
        fungsiminuman()
        lagi = input("Apakah Anda ingin memesan lagi? (ya/tidak): ").lower()
        if lagi != "ya":
            break

menu_ulang()

totalsemua = totalmkn + totalmnm
totalsemua = hitung_diskon(totalsemua)

print("\nTotal Yang Harus Dibayar:", totalsemua)
uang = int(input("Uang Tunai Pembeli: "))
kembalian = uang - totalsemua
print("Uang Kembalian:", kembalian)

print("\n=================================")
print("===========STRUK BELI============")
print("=================================")
print("Nama\t\t:", pembeli)
print("Tagihan\t\t: Rp", totalsemua)
print("Dibayar\t\t: Rp", uang)
print("Kembalian\t: Rp", kembalian)
print("=================================")
print("Terima Kasih atas kunjungan Anda!")
```
